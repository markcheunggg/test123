## Boolean Indexing

We can use the comparison operator **nba_guard_stat\[\'GP\'\] >= 30** to find all the players who have played at least 30 games.

**Boolean Indexing:**
dataframe\[column = n\]

**Example:**
nba_guard_stats[nba_guard_stats['GP'] >= 30]

Boolean indexing allows us to throw out all the rows that are false and keep all the rows that are true.


```python
#Any row that does not satisfy the comparison operator is removed from the table!
cleaned_nba_guard_stats = nba_guard_stats[nba_guard_stats['GP'] >= 30]
cleaned_nba_guard_stats
```

## Indexing (iloc and iloc)
You'll use these a lot. Super important! **iloc** and **loc** allow you to locate rows, columns, and single datapoints in a dataset.

### Integer Location (iloc)

Think of "iloc" as "integer location". It uses the indexes of rows and columns to find certain datapoints.

**iloc\[*row, column*\]** will find the datapoint at the specified row and column!

ex: **iloc\[*5, 10*\]** will find the datapoint at the 5th row and 10th column.


```python
cleaned_nba_guard_stats.head(5)
```

```python
#Element at the 0th row and 0th column
cleaned_nba_guard_stats.iloc[0,0]
```

To get a single row using **iloc**, you can do **iloc\[*row index*\]**

```python
#James Harden's row
cleaned_nba_guard_stats.iloc[0]
```

To get a single column using **iloc**, you can do **iloc\[*:, column index*\]**

```python
#Get the first 10 players in the player column.
cleaned_nba_guard_stats.iloc[:10, 0]
```

### Location (loc)
Think of **loc** as *label-based* location. It uses the names of columns to find datapoints.

It's also row-first, column-second:
**loc\[*row, column*\]**

ex: **iloc\[*5, "PLAYER"*\]** will find the 5th row in the "PLAYER" column.


```python
cleaned_nba_guard_stats.head(5)
```

```python
#Get the element in the "PLAYER" column at the 0th row index. 
cleaned_nba_guard_stats.loc[0,"PLAYER"]
```

```python
#Get the first 10 players in the "PLAYER" column
cleaned_nba_guard_stats.loc[:10,"PLAYER"]
```

```python
#Get James Harden's Row
cleaned_nba_guard_stats.loc[0]
```

Like iloc, the row identifier will be an integer (the index) because rows are labelled by their index.
However, you can change the index to a column so that you can use a more specific label to identify the rows!

```python
#Changes the index column to the player column
cleaned_nba_guard_stats_changed_index = cleaned_nba_guard_stats.set_index("PLAYER")
cleaned_nba_guard_stats_changed_index.head(5)
```

```python
#Now you can use a Player's Name as the row index to locate a player.
#The following line looks for "Russell Westbrook" row in the index and for the PTS column.
cleaned_nba_guard_stats_changed_index.loc["Russell Westbrook","PTS"]
```

## Practical example
Imagine you're doing an analysis of archetypes (types of skillsets) amongst NBA guards. 
Let's say you define a **"three-and-D" player** (a player who excels at playing defense and shooting 3PTs) as a player who shoots **at least 38.0% on 3PTs** and averages **at least 1.0 steals** per game.

```python
#Uses boolean indexing to get all the players who shoot above 38.0% 3PT% and average above 1.0 steals per game.
three_and_d = cleaned_nba_guard_stats[(cleaned_nba_guard_stats['3P%'] >= .38) & (cleaned_nba_guard_stats['STL'] >= 1.0)]

#Uses loc to get the player column.
three_and_d_players = three_and_d.loc[:, ["PLAYER", "STL", "3P%"]]
three_and_d_players
```

```python
#If you want it as an array:
three_and_d_players.to_numpy()
```

## Math operations
You can also add, substract, multiply, and divide columns!

```python
cleaned_nba_guard_stats.head(5)
```

```python
#How to find steals per minute: Steals column divided by minutes column 
per_1_min__steals = cleaned_nba_guard_stats["STL"] / cleaned_nba_guard_stats["MIN"]
per_36_min_steals = round(per_1_min__steals * 36, 1)

per_36_min_steals_table = cleaned_nba_guard_stats
per_36_min_steals_table["STL per 36"] = per_36_min_steals
per_36_min_steals_table.head(5)
```

```python
three_and_d = cleaned_nba_guard_stats[(cleaned_nba_guard_stats['3P%'] >= .38) & (cleaned_nba_guard_stats['STL per 36'] >= 1.0)]
three_and_d_players = three_and_d.loc[:, ["PLAYER", "TEAM", "STL per 36", "3P%"]]
three_and_d_players
```

## Visualize your data! (Seaborn)
Using Seaborn, you can create data visualizations using the columns of your pandas table!

**sns.scatterplot(*x = x-axis data, y = y-axisdata*)**

```python
sns.scatterplot(x=three_and_d_players['STL per 36'], y=three_and_d_players['3P%'])
```

### Color-coding data
You can color-code your data by a certain column using the **hue** parameter!

**sns.scatterplot(*x = x-axis data, y = y-axisdata, hue = column you want to color-code by*)**

```python
#Color-codes the graph by team.
sns.scatterplot(x=three_and_d_players['STL per 36'], y=three_and_d_players['3P%'], hue=three_and_d_players['TEAM'])
```

```python
#Gets three-pointers attempted per 36 minutes
per_1_min__3PA = cleaned_nba_guard_stats["3PTA"] / cleaned_nba_guard_stats["MIN"]
per_36_min_3PA = round(per_1_min__3PA * 36, 1)

per_36_min = per_36_min_steals_table
per_36_min["3PA per 36"] = per_36_min_3PA

per_36_min = per_36_min.loc[:, ["PLAYER", "TEAM", "3P%", "STL per 36",  "3PA per 36"]]
per_36_min.head(5)
```

## Adjusting size of data-points
The **size** parameter allows you to compare you data with a 3rd factor!

**sns.scatterplot(*x = x-axis data, y = y-axisdata, size = column you want size to be based on*)**

```python
#Will make the dots a different size depending on the volume of 3PA attempted by each player.
sns.scatterplot(x=per_36_min['STL per 36'], y=per_36_min['3P%'], size=per_36_min["3PA per 36"])
```

## Actual Analysis! (Linear Regressions)


```python
college_shooting_stats = pd.read_csv("College Shooting Stats.csv")
college_shooting_stats.head(5)
```

```python
sns.scatterplot(x=college_shooting_stats["3P% College"], y=college_shooting_stats["3P% 1st Season"])
```

```python
sns.scatterplot(x=college_shooting_stats["FT% College"], y=college_shooting_stats["3P% 1st Season"])
```

```python
three_point_predictor = college_shooting_stats["3P% College"]
free_throw_predictor = college_shooting_stats["FT% College"]
rookie_three_point = college_shooting_stats["3P% 1st Season"]
```

```python
three_point_predictor.corr(rookie_three_point)
```

```python
free_throw_predictor.corr(rookie_three_point)
```

```python
model = np.poly1d(np.polyfit(free_throw_predictor, rookie_three_point, 1))
```

```python
plt.scatter(free_throw_predictor, rookie_three_point)
plt.plot(np.linspace(.5, 1, 50), model(np.linspace(.5, 1, 50)))
```

```python
rookie_shooting_stats = pd.read_csv("Rookie Shooting Stats.csv")
rookie_shooting_stats.head(5)
```

```python
predicted_three_point = model(rookie_shooting_stats["FT% College"])
predicted_three_point
```

```python
rookie_shooting_stats["Predicted 3P%"] = predicted_three_point
rookie_shooting_stats
```
